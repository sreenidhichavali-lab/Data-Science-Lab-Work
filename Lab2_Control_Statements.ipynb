```python
### indentation:
### control statement
## to check if a number is positive,negative or zero

num=int(input("Enter your number: "))
if num>0:
  print("The number is positive")

elif num<0:
  print("The number is negative")
elif num==0:
  print("The number is zero")
```

```python
## whether the give number is odd even

num=int(input("Enter your number: "))
if num%2==1:
  print("The number is odd")
elif num%2==0:
  print("The number is even")
```

```python
### python code to print multiplication table

for i in range(1,11):
  for j in range(1,11):
    print("{:5}".format(i*j), end = ' ')
  print()
```

```python
### while loop

count = 0
while(count<5):
  print("My count is ", count)
  count = count+1
```

```python
Students= ("x1","x2","x3","x4","x5","x6")

for i in Students:
  if len(i)>6:
    print(i)
```

```python
### enumerate
Students= ("x1","x2","x3","x4","x5","x6")
for ind, stu in enumerate(Students):
  print(ind, stu)
```

```python
### Loop in the dictionary

student = {"x1": 82, "x2": 85,"x3": 90, "x4": 91, "x5": 55}
for i in student:
  marks=student[i]
  #print(marks)
  print("Student %s has scored %d " %(i,marks))
```

```python
student.values()
```

```python
for i in student.values():
  print(i)
```

```python
for k, v in student.items():
  print(k,v)
```
