container types in python
The four core built-in container types are:
list: A mutable, ordered sequence of elements. Lists are indexed by integers and can hold items of different types. They are defined using square brackets (e.g., [1, "a", 2.5]).
tuple: An immutable, ordered sequence of elements. Tuples are similar to lists but cannot be changed after creation. They are defined using parentheses (e.g., (1, "a", 2.5)).
dict (Dictionary): A mutable mapping type that stores data as unordered (ordered since Python 3.6) key-value pairs. Data is accessed using unique keys, not integer indices. They are defined using curly braces (e.g., {"name": "John", "age": 30}).
set: A mutable, unordered collection of unique elements. Sets are useful for operations like union and intersection and for quickly checking membership. They are defined using curly braces (e.g., {1, 2, 3}) or the set() constructor.

Confusion Matrix Metrics (Python as Calculator)               
*   **Problem:** A binary classification model predicts whether rainfall will cause Flood (Yes/No).




```python
## Given Values
TP = 50
FP = 8
TN = 32
FN = 10
```

```python
## 1. Total Predictions
total_predictions = TP + FP + TN + FN
print("Total Predictions:", total_predictions)
```

```python
## 2. Accuracy
accuracy = (TP + TN) / total_predictions
print("Accuracy:", accuracy)
```

```python
## 3. Precision
precision = TP / (TP + FP)
print("Precision:", precision)
```

```python
## 4. Recall
recall = TP / (TP + FN)
print("Recall:", recall)
```

```python
## F1 Score
f1_score = 2 * (precision * recall) / (precision + recall)
print("F1 Score:", f1_score)
```

In a regression model predicting river discharge, the following are known:
*   Sum of Squares due to  Regression (SSR) = 420
*   Sum of Squares due to Error (SSE) = 180




```python
## Given Values

SSR = 420
SSE = 180
```

```python
## Total Sum of Squares
SST=SSR+SSE
print("Total Sum of squares (SST):", SST)
```

```python
## R-squared calculation
R_squared =SSR/SST
print("R-squared (R²):", R_squared)
```

```python
## Interpretation
print("Model explains more than 70% variance:", R_squared > 0.7)
```




Problem 4: Cook's Distance(Influential Observation)

For a regression model:

*  Squared residual (eᵢ²) = 9
*  Mean Squared Error (MSE) = 3
*  Number of predictors (p) = 3
*  Leverage value (hᵢᵢ) = 0.2
*  Sample size(n)=25

```python
## given values
ei_squared=9
MSE =3
p=3
hii=0.2
n=25
```

```python
## Cook's Distance formula
cooks_distance=(ei_squared / (p*MSE)) *(hii/(1-hii)**2)
print("Cook's Distance:", cooks_distance)
```

```python
## threshold
threshold=4/n
print("Threshold:", threshold)
```

```python
print("Is the observation influential:", cooks_distance>threshold)
```

  
Problem 5: Simple Linear Regression Prediction (Calculator Style)
*   A regression model for predicting crop yield is:
y=2.5x+10

Given rainfall value:
	x=12


```python
## Given values
slope=2.5
x=12
c=10
```

```python
predicted_yield= (slope * x)+c
print("Predicted Crop Yield:", predicted_yield)
```

Slope (2.5): Indicates that for every 1 unit increase in rainfall, the crop yield increases by 2.5 units.

Intercept (10): Represents the expected crop yield when rainfall is zero.
